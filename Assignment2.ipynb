
# Python Set Comprehensions

Python set comprehensions are similar to list comprehensions and allow a programmer to quickly create sets using simple, concise, and readable code.

To demonstrate Set Comprhensions we are going to manipulate and work with Countries. To do this we will use the pycountry library.

## Step #2

#### Next we will import pycountry into the project and print a count of how many countries are in the list



```python
import pycountry

len(pycountry.countries)
```

## Step #3

### Next we will create a list of all 249 countries and print the first one to show its format

```python
countries = list(pycountry.countries)
countries[1]
```



## Step #4

### Next we create a list containing every word used in any of the official_name attributes from the list of countries

```python
official_names = []
for x in countries:        
    if hasattr(x, 'official_name'):
        official_names.extend(x._fields['official_name'].split())
official_names
```



## Step #4

### Last, we will create a set containing all unique words used to form a countries official name.

```python
a = {x for x in official_names}
a
```
