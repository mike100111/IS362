## Step #1
### Import Pandas and data

```python
import pandas as pd
import numpy as np
import seaborn as sns

data = pd.read_csv('auto-mpg.data', delim_whitespace=True)
data.head()
```

## Step #2
### Add column names

```python
data.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin', 'car name']
data.head()
```

## Step #3
### Replace ? with NaN and convert origin from int to text description

```python
data = data.replace('?', np.nan)
data[['origin']] = data[['origin']].replace([1, 2,3],['USA','Asia','Europe'])
data.head()
```

## Step #4
### Display a bar chart by origin

```python
sns.barplot(x=data.cylinders, y=data.origin)
```

## Step #5 
### Display data on a scatter chart

```python
sns.scatterplot(x=data.weight, y=data.horsepower)
```

## Step #6
### compare data by origin by coloring data points to see which origin has the heaviest and most horsepower.

```python
sns.scatterplot(x=data.weight, y=data.horsepower, hue=data.origin)
```
